```python
df=df.sort_values(by='DATE',ascending=True).reset_index(drop=True)

df.head()
```

```python
df.columns=df.columns.str.lower()

df.columns
```

```python
df['date']=df['date'].astype('datetime64[ns]')

print(df.dtypes['date'])
```

```python
df['gprd_pct_change']        = df['gprd'].pct_change(fill_method=None) * 100 
df['gprd_act_pct_change']    = df['gprd_act'].pct_change(fill_method=None) * 100
df['gprd_threat_pct_change'] = df['gprd_threat'].pct_change(fill_method=None) * 100
df['gprd_shock'] = df['gprd_pct_change'].abs()
```

```python
df.head(15)
```

```python
df[['gold_change_%','silver_change_%','gprd_pct_change']].describe()
```

```python
price_cols = ['gold_price', 'silver_price']

df_prices = df.melt(
    id_vars='date',
    value_vars=price_cols,
    var_name='Metal',
    value_name='Price'
)

fig = px.line(
    df_prices,
    x='date',
    y='Price',
    color='Metal',
    title='Precios de oro y plata a lo largo del tiempo'
)
fig.show()
```

```python
risk_cols = ['gprd', 'gprd_act', 'gprd_threat']

df_risk = df.melt(
    id_vars='date',
    value_vars=risk_cols,
    var_name='Index',
    value_name='Value'
)

fig = px.line(
    df_risk,
    x='date',
    y='Value',
    color='Index',
    title='Evolucin del riesgo geopoltico'
)
fig.show()
```

```python
fig = px.scatter(
    df,
    x='gprd_pct_change',      # cambio % en el ndice
    y='gold_change_%',        # cambio % en el oro
    title='Cambio % en GPRD vs cambio % en el oro',
    labels={
        'gprd_pct_change': 'Cambio % diario GPRD',
        'gold_change_%': 'Cambio % diario oro'
    }
)
fig.show()
```

```python
df_ret = df.melt(
    id_vars='date',
    value_vars=['gold_change_%', 'silver_change_%'],
    var_name='Metal',
    value_name='Return_%'
)

fig = px.box(
    df_ret,
    x='Metal',
    y='Return_%',
    title='Distribucin de cambios % diarios en oro y plata'
)
fig.show()
```
